```python
import pandas as pd
```

```python
old_cars = pd.read_csv('old_cars.csv')
new_cars = pd.read_csv('new_cars.csv')
```

```python
old_cars.head()

# drop the first column
old_cars.drop(old_cars.columns[0], axis=1, inplace=True)

old_cars.head()
```

```python
new_cars.head()
```

```python
# merge the two dataframes with new column 'new' to indicate if the car is new or not

old_cars['new'] = 0
new_cars['new'] = 1

cars = pd.concat([old_cars, new_cars], ignore_index=True)

cars.head()
```

```python
cars.count()
```

```python
# save the dataframe to a csv file
cars.to_csv('merged_cars.csv', index=False)
```
